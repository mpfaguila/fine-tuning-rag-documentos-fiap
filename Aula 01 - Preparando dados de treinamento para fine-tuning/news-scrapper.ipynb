```python
#Definindo a função que retornará os links

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def scrape_cnn_links(url="https://edition.cnn.com/world"):
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/124.0 Safari/537.36"
        )
    }
    resp = requests.get(url, headers=headers, timeout=20)
    if resp.status_code != 200:
        print(f"Falha ao acessar a página: {resp.status_code}")
        return []

    soup = BeautifulSoup(resp.text, "html.parser")

    candidates = []
    for a in soup.select("a[href]"):
        href = a.get("href")
        if not href:
            continue
        # Torna a URL absoluta
        full = urljoin(url, href)
        # Heurística de matérias (CNN costuma ter /202X/… em notícias)
        if "/202" in full and "cnn.com" in full:
            candidates.append(full)

    # Remove duplicatas e itens não-HTML comum (evita âncoras)
    links = sorted({u.split("?")[0] for u in candidates if u.startswith("http")})
    return links

# Uso e gravação:
links = scrape_cnn_links("https://edition.cnn.com/world")
with open("CNN_Links.txt", "w", encoding="utf-8") as f:
    for link in links:
        f.write(link + "\n")

print(f"{len(links)} links salvos em CNN_Links.txt")
```
